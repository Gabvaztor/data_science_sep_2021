```python
def recorre_diccionario_3(diccionario):
    # recorre por clave y values y los muestra
    for key, value in diccionario.items():
        print("---------")
        print("key:", key)
        print("value:", value)

recorre_diccionario_3(diccionario=d)
```

```python
def recorre_diccionario_4(diccionario):
    # recorre por clave y values y los muestra
    for item in diccionario.items():
        print("---------")
        print("item:", item)
        print("key:", item[0])
        print("value:", item[1])

recorre_diccionario_4(diccionario=d)
```

```python
def recorre_diccionario_5(diccionario):
    # Crea una lista de tuplas del diccionario
    # Al final, muestra la lista de tuplas por pantalla
    lista = []
    for item in diccionario.items():
        lista.append(item)
    print("lista de tuplas:", lista)

recorre_diccionario_5(diccionario=d)
```

```python
def recorre_diccionario_6(diccionario):
    # Crea una lista de tuplas del diccionario
    # Al final, retorna la lista de tuplas
    lista = []
    for item in diccionario.items():
        lista.append(item)
    return lista

recorre_diccionario_6(diccionario=d)
```

```python
# while

suma_acumulada = 0
acumulador = 0
while acumulador <= 2:
    numero_usuario = int(input("Inserte un número:"))
    suma_acumulada = suma_acumulada + numero_usuario
    print("numero_usuario:", numero_usuario)
    print("suma_acumulada:", suma_acumulada)
    acumulador = acumulador + 1
    print("acumulador:", acumulador)

print("FIN")
```

```python
# while indefinido hasta escribir el número 99999

suma_acumulada = 0
acumulador = 0
while acumulador >= 0:
    numero_usuario = int(input("Inserte un número:"))
    if numero_usuario == 99999:
        print("Se para el while")
        break
    suma_acumulada = suma_acumulada + numero_usuario
    print("numero_usuario:", numero_usuario)
    print("suma_acumulada:", suma_acumulada)
    acumulador = acumulador + 1
    print("acumulador:", acumulador)

print("FIN")
```

```python
# Listas con tuplas

lista_tupla = [("A", "X", -99, "Y"), (2, 4, 6, 8)]

# Eliminar la tupla que contenga el int 4
def elimina_tupla_que_contenga_4(arg):
    for tupla in arg:
        if 4 in tupla:
            arg.remove(tupla)
    print("La lista modificada:", arg)


elimina_tupla_que_contenga_4(arg=lista_tupla)
```

```python
# Listas con tuplas

lista_tupla = [(2, 4, 6, 8), ("A", "X", -99, "Y"), (2, 4, 6, 8)]

# Eliminar la tupla que contenga el int 4
def elimina_tupla_que_contenga_4(arg):
    lista_de_valores_a_eliminar = []
    # Creo la lista de los valores que voy a eliminar
    for tupla in arg:
        if 4 in tupla:  # Está en 4 en la tupla
            lista_de_valores_a_eliminar.append(tupla)
    print("lista_de_valores_a_eliminar:", lista_de_valores_a_eliminar)

    # Recorro los valores a eliminar y los elimino de la lista original
    for valor_a_eliminar in lista_de_valores_a_eliminar:
        arg.remove(valor_a_eliminar)

    print("La lista modificada:", arg)


elimina_tupla_que_contenga_4(arg=lista_tupla)
```

- Dentro de una función si el argumento es una lista:

    1. Si modificas la lista internamente (no le cambias el tipo lista) --> los cambios son permanentes a la variable fuera de la función
    2. Si le asignas un tipo diferente al argumento, la variable externa a la función no cambiará de valor


```python
x = [3,5]

def cambia_x(arg):
    arg.pop()
    arg = 6
    print("arg:", arg)

cambia_x(arg=x)
print("x:", x)
```
